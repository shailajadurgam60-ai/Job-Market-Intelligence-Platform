```python
def create_skill_dataframe(skill_count):

    sorted_skills = sorted(
        skill_count.items(),
        key=lambda x:x[1],
        reverse=True
    )

    return pd.DataFrame(
        sorted_skills,
        columns=["skill","count"]
    )
```

```python
def filter_role(dataframe, role):

    return dataframe[
        dataframe["title"].str.contains(
            role,
            case=False,
            na=False
        )
    ]
```

```python
def plot_bar_chart(dataframe, x_column, y_column, title,top_n=None):
    if top_n is not None:
      dataframe = dataframe.nlargest(top_n, x_column)
    plt.figure(figsize=(10,6))

    ax = sns.barplot(
        data=dataframe,
        x=x_column,
        y=y_column,
        color="steelblue"
    )

    plt.title(title, fontsize=16, fontweight="bold")

    plt.xlabel(x_column.capitalize())
    plt.ylabel(y_column.capitalize())

    for container in ax.containers:
        ax.bar_label(container, padding=5)

    plt.grid(axis="x", linestyle="--", alpha=0.5)

    plt.xlim(0, dataframe[x_column].max() * 1.1)

    plt.tight_layout()

    plt.show()
```

```python
plot_bar_chart(
    analyst_df,
    "count",
    "skill",
    "Top 10 Skills for Data Analyst",
    top_n=10
)
plot_bar_chart(
    scientist_df.nlargest(10, "count"),
    "count",
    "skill",
    "Top 10 Skills for Data Scientist",
    top_n=10
)
plot_bar_chart(
    engineer_df.nlargest(10, "count"),
    "count",
    "skill",
    "Top 10 Skills for Data Engineer",
    top_n=10
)
```

#Insights from Role-wise skill Analysis

1.Overall Job Market
  * SQL and Python are the two most frequently requested skills across all data-related roles.
   * Cloud platforms like AWS and Azure are increasingly common, showing the industry's move toward cloud-  based data solutions.
  * Visualization tools such as Tableau and Power BI remain important, especially for analytical roles.

2.Data Analyst
* SQL is the most demanded skill, indicating that querying and analyzing structured data is a core responsibility.
* Excel remains highly relevant, reflecting its continued use in business reporting and analysis.
* Tableau and Power BI emphasize the importance of communicating insights through dashboards.
* Python appears frequently, showing that programming skills are becoming increasingly valuable for analysts.
  
3.Data Scientist
* Python is the dominant skill, highlighting its role in data analysis and machine learning.
* Machine Learning and Statistics are among the most requested skills, reflecting the predictive nature of the role.
* SQL continues to be essential for accessing and preparing data.
* Spark appears more frequently than in analyst roles, indicating the need to work with larger datasets.

4.Data Engineer
* SQL and Python form the foundation of Data Engineering.
* Spark, Hadoop, and Kubernetes indicate strong demand for big data and distributed computing technologies.
* AWS and Azure highlight the importance of cloud platforms in modern data engineering workflows.
* Compared with analysts and data scientists, infrastructure-related technologies are more prominent.

  | Observation     | Finding                                                |

| Analyst focus   | Business intelligence, reporting, and visualization.   |
