```python
masked_companies = df["company"].str.contains(r"\*", regex=True, na=False).sum()
masked_titles = df["title"].str.contains(r"\*", regex=True, na=False).sum()
masked_locations = df["location"].str.contains(r"\*", regex=True, na=False).sum()

print("Masked companies:", masked_companies)
print("Masked titles:", masked_titles)
print("Masked locations:", masked_locations)
```
